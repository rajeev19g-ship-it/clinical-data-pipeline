```
         |
         v

         |
         v
Normalisation         --  Library size normalisation · Log transformation
         |
         v
Feature Selection     --  Highly variable genes · PCA
         |
         v

         |
         v
Clustering            --  Leiden · Louvain
         |
         v
Cell Type Classification  --  Marker genes · ML classifier
         |
         v
Differential Expression   --  DEGs per cell type · Pathway enrichment
```

Here we simulate a realistic PBMC dataset with known cell type composition
including neutrophils — the key cell type of interest in the Takeda programme.

````python
np.random.seed(42)

N_CELLS  = 8000
N_GENES  = 2000

# Cell type composition — realistic PBMC proportions
# Neutrophils are typically rare in standard PBMC preps
# The Takeda pipeline improved their recovery to ~80%
CELL_TYPES = {
    'CD4+ T Cell':       0.28,
    'CD8+ T Cell':       0.18,
    'B Cell':            0.12,
    'NK Cell':           0.10,
    'CD14+ Monocyte':    0.12,
    'CD16+ Monocyte':    0.06,
    'Dendritic Cell':    0.04,
    'Neutrophil':        0.06,   # Enhanced recovery vs standard ~2%
    'Regulatory T Cell': 0.02,
    'Plasmablast':       0.02,
}

# Canonical marker genes per cell type
MARKER_GENES = {
    'CD4+ T Cell':       ['CD3D', 'CD3E', 'CD4', 'IL7R', 'CCR7', 'SELL'],
    'CD8+ T Cell':       ['CD3D', 'CD8A', 'CD8B', 'GZMB', 'PRF1', 'NKG7'],
    'B Cell':            ['CD79A', 'CD79B', 'MS4A1', 'CD19', 'IGHM', 'IGLC2'],
    'NK Cell':           ['GNLY', 'NKG7', 'KLRD1', 'NCAM1', 'FCGR3A', 'GZMB'],
    'CD14+ Monocyte':    ['CD14', 'LYZ', 'S100A8', 'S100A9', 'CST3', 'VCAN'],
    'CD16+ Monocyte':    ['FCGR3A', 'MS4A7', 'CDKN1C', 'HES4', 'LILRA3'],
    'Dendritic Cell':    ['FCER1A', 'CST3', 'HLA-DQA1', 'CLEC10A', 'CD1C'],
    'Neutrophil':        ['FCGR3B', 'CSF3R', 'CXCR2', 'S100A8', 'MMP8', 'ELANE', 'MPO'],
    'Regulatory T Cell': ['CD3D', 'CD4', 'FOXP3', 'IL2RA', 'CTLA4', 'TIGIT'],
    'Plasmablast':       ['IGHG1', 'MZB1', 'SDC1', 'JCHAIN', 'PRDM1'],
}

# Assign cell types
cell_type_list = list(CELL_TYPES.keys())
cell_type_probs = list(CELL_TYPES.values())
cell_labels = np.random.choice(cell_type_list, size=N_CELLS, p=cell_type_probs)

# Generate sparse count matrix with cell-type-specific expression patterns
gene_names = [f'GENE_{i:04d}' for i in range(N_GENES)]

# Add marker genes
all_markers = []
for markers in MARKER_GENES.values():
    all_markers.extend(markers)
all_markers = list(set(all_markers))
gene_names[:len(all_markers)] = all_markers

# Build count matrix
counts = np.random.negative_binomial(1, 0.8, size=(N_CELLS, N_GENES)).astype(np.float32)

# Boost marker gene expression per cell type
for cell_type, markers in MARKER_GENES.items():
    mask = cell_labels == cell_type
    for marker in markers:
        if marker in gene_names:
            idx = gene_names.index(marker)
            counts[mask, idx] += np.random.negative_binomial(8, 0.3, size=mask.sum())

# Library size (UMI counts per cell) — realistic 10x distribution
library_sizes = np.random.lognormal(mean=8.5, sigma=0.5, size=N_CELLS).astype(int)
library_sizes = np.clip(library_sizes, 500, 50000)

print(f'Simulated 10x dataset:')
print(f'  Cells:     {N_CELLS:,}')
print(f'  Genes:     {N_GENES:,}')
print(f'  Sparsity:  {(counts == 0).mean():.1%}')
print(f'  Median UMI: {np.median(library_sizes):,.0f}')
print(f'\nCell type distribution:')
for ct, count in pd.Series(cell_labels).value_counts().items():
    pct = count / N_CELLS * 100
    print(f'  {ct:<25} {count:>5} cells ({pct:.1f}%)')
````

## 3. Quality Control

Standard 10x QC metrics:
- **n_genes_by_counts**: Number of genes detected per cell
- **total_counts**: Total UMI counts per cell  
- **pct_counts_mt**: Percentage mitochondrial gene counts (high = dead/damaged cells)

The Takeda pipeline applied adaptive thresholding rather than fixed cutoffs,
which was critical for achieving 80% cell recovery — fixed thresholds were
discarding viable neutrophils due to their naturally lower RNA content.

````python
# Compute QC metrics
n_genes_per_cell   = (counts > 0).sum(axis=1)
total_counts_cell  = counts.sum(axis=1)
pct_mt             = np.random.beta(1.5, 20, N_CELLS) * 100  # Simulated MT %

# Neutrophils have naturally lower gene counts — this is biology not poor quality
neutrophil_mask = cell_labels == 'Neutrophil'
pct_mt[neutrophil_mask] = np.random.beta(1, 25, neutrophil_mask.sum()) * 100

# ── Adaptive QC thresholds ────────────────────────────────────────────────────
# Key insight: fixed thresholds (e.g. n_genes > 200) discard neutrophils
# because they have fewer expressed genes than lymphocytes.
# Adaptive thresholds based on median absolute deviation (MAD) preserve them.

def mad_threshold(values, n_mads=5, direction='both'):
    """Adaptive threshold using Median Absolute Deviation."""
    median = np.median(values)
    mad    = np.median(np.abs(values - median))
    low    = median - n_mads * mad if direction in ['both', 'low']  else -np.inf
    high   = median + n_mads * mad if direction in ['both', 'high'] else np.inf
    return low, high

# Standard fixed thresholds (would lose neutrophils)
fixed_qc = (
    (n_genes_per_cell   >= 200) &
    (n_genes_per_cell   <= 6000) &
    (total_counts_cell  >= 500) &
    (pct_mt             <= 20)
)

# Adaptive MAD-based thresholds (preserves rare cell types)
gene_low,  gene_high  = mad_threshold(np.log1p(n_genes_per_cell), n_mads=5)
count_low, count_high = mad_threshold(np.log1p(total_counts_cell), n_mads=5)
mt_low,    mt_high    = mad_threshold(pct_mt, n_mads=3, direction='high')

adaptive_qc = (
    (np.log1p(n_genes_per_cell)  >= gene_low)  &
    (np.log1p(n_genes_per_cell)  <= gene_high) &
    (np.log1p(total_counts_cell) >= count_low) &
    (pct_mt <= mt_high)
)

# Recovery comparison
fixed_neutrophil_recovery    = (fixed_qc    & neutrophil_mask).sum() / neutrophil_mask.sum()
adaptive_neutrophil_recovery = (adaptive_qc & neutrophil_mask).sum() / neutrophil_mask.sum()

print('=== Quality Control Comparison ===')
print(f'\nFixed thresholds:')
print(f'  Total cells retained:      {fixed_qc.sum():,} / {N_CELLS:,} ({fixed_qc.mean():.1%})')
print(f'  Neutrophil recovery:       {fixed_neutrophil_recovery:.1%}')
print(f'\nAdaptive MAD thresholds (Takeda approach):')
print(f'  Total cells retained:      {adaptive_qc.sum():,} / {N_CELLS:,} ({adaptive_qc.mean():.1%})')
print(f'  Neutrophil recovery:       {adaptive_neutrophil_recovery:.1%}')
print(f'\nImprovement in neutrophil recovery: +{(adaptive_neutrophil_recovery - fixed_neutrophil_recovery)*100:.1f}%')

# Apply adaptive QC
counts_qc       = counts[adaptive_qc]
cell_labels_qc  = cell_labels[adaptive_qc]
n_cells_qc      = counts_qc.shape[0]

print(f'\nPost-QC dataset: {n_cells_qc:,} cells retained')
````

## 4. Normalisation and Feature Selection

````python
# Library size normalisation — scale each cell to 10,000 total counts
counts_norm = counts_qc / counts_qc.sum(axis=1, keepdims=True) * 10000

# Log transformation — stabilises variance
counts_log = np.log1p(counts_norm)

# Highly Variable Gene (HVG) selection
# Select genes with highest variance — informative for cell type separation
gene_means  = counts_log.mean(axis=0)
gene_vars   = counts_log.var(axis=0)
gene_cv2    = gene_vars / (gene_means ** 2 + 1e-8)  # Coefficient of variation squared

n_hvg       = 500
hvg_indices = np.argsort(gene_cv2)[::-1][:n_hvg]
counts_hvg  = counts_log[:, hvg_indices]

# PCA for dimensionality reduction
n_pcs  = 50
scaler = StandardScaler()
counts_scaled = scaler.fit_transform(counts_hvg)

pca    = PCA(n_components=n_pcs, random_state=42)
pcs    = pca.fit_transform(counts_scaled)

explained_var = pca.explained_variance_ratio_.cumsum()

print(f'Normalisation complete:')
print(f'  Highly variable genes selected: {n_hvg}')
print(f'  PCA components: {n_pcs}')
print(f'  Variance explained (PC1-10):  {explained_var[9]:.1%}')
print(f'  Variance explained (PC1-50):  {explained_var[49]:.1%}')

# Determine optimal number of PCs using elbow method
pc_diffs    = np.diff(pca.explained_variance_ratio_)
optimal_pcs = np.argmin(pc_diffs[:20]) + 2
print(f'  Optimal PCs (elbow):          {optimal_pcs}')

# Use optimal PCs for downstream analysis
pcs_use = pcs[:, :optimal_pcs]
````

## 6. ML Cell Type Classifier

Automated cell type classification using gradient boosting on PCA embeddings.
This replaces manual marker-gene annotation — critical for scalability when
processing thousands of samples in a clinical pipeline.

````python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, balanced_accuracy_score

# Encode cell type labels
le = LabelEncoder()
y  = le.fit_transform(cell_labels_qc)

# Train/test split — stratified to ensure rare cell types in test set
X_train, X_test, y_train, y_test = train_test_split(
    pcs_use, y,
    test_size=0.2,
    random_state=42,
    stratify=y,
)

# Gradient Boosting classifier
# Key parameter: class_weight balanced — critical for rare cell types (neutrophils)
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(
    n_estimators=200,
    max_depth=15,
    class_weight='balanced',    # Handles class imbalance for rare cell types
    random_state=42,
    n_jobs=-1,
)

clf.fit(X_train, y_train)
y_pred  = clf.predict(X_test)
y_proba = clf.predict_proba(X_test)

# Overall metrics
balanced_acc = balanced_accuracy_score(y_test, y_pred)

print('=== Cell Type Classification Results ===')
print(f'\nBalanced Accuracy: {balanced_acc:.4f}')
print(f'\nPer-class metrics:')
print(classification_report(
    y_test, y_pred,
    target_names=le.classes_,
    digits=3,
))

# Neutrophil-specific metrics
neut_idx  = list(le.classes_).index('Neutrophil')
neut_mask = y_test == neut_idx
neut_pred = y_pred[neut_mask]
neut_recall = (neut_pred == neut_idx).sum() / neut_mask.sum()

print(f'\nNeutrophil detection metrics:')
print(f'  Recall (sensitivity):  {neut_recall:.3f}')
print(f'  Cells in test set:     {neut_mask.sum()}')
print(f'  Correctly identified:  {(neut_pred == neut_idx).sum()}')
````

## 7. Differential Expression Analysis

Identify marker genes and differentially expressed genes (DEGs) per cell type.
Used at Takeda for drug target identification and biomarker discovery.

````python
from scipy import stats

def find_marker_genes(
    counts_matrix: np.ndarray,
    cell_labels: np.ndarray,
    gene_names: list,
    target_cell_type: str,
    n_top_genes: int = 20,
    min_pct: float = 0.1,
) -> pd.DataFrame:
    """
    Find marker genes for a cell type using Wilcoxon rank-sum test.

    Identifies genes significantly upregulated in the target cell type
    vs all other cells. This is the standard approach for automated
    cell type annotation in clinical scRNA-seq pipelines.

    Parameters
    ----------
    target_cell_type : str
        Cell type to find markers for.
    min_pct : float
        Minimum fraction of target cells expressing the gene.
    """
    target_mask = cell_labels == target_cell_type
    other_mask  = ~target_mask

    target_expr = counts_matrix[target_mask]
    other_expr  = counts_matrix[other_mask]

    results = []
    for i in range(min(500, counts_matrix.shape[1])):
        target_vals = target_expr[:, i]
        other_vals  = other_expr[:, i]

        # Expression fraction filter
        pct_target = (target_vals > 0).mean()
        pct_other  = (other_vals  > 0).mean()

        if pct_target < min_pct:
            continue

        # Wilcoxon rank-sum test
        try:
            stat, pval = stats.ranksums(target_vals, other_vals)
        except Exception:
            continue

        # Log2 fold change
        mean_target = target_vals.mean()
        mean_other  = other_vals.mean()
        log2fc = np.log2((mean_target + 1e-6) / (mean_other + 1e-6))

        results.append({
            'gene':       gene_names[i] if i < len(gene_names) else f'GENE_{i}',
            'log2fc':     round(log2fc, 3),
            'pval':       pval,
            'pct_target': round(pct_target, 3),
            'pct_other':  round(pct_other, 3),
            'mean_target': round(mean_target, 3),
        })

    if not results:
        return pd.DataFrame()

    df = pd.DataFrame(results)

    # Bonferroni correction
    from statsmodels.stats.multitest import multipletests
    try:
        _, pval_adj, _, _ = multipletests(df['pval'], method='bonferroni')
        df['pval_adj'] = pval_adj
    except ImportError:
        df['pval_adj'] = df['pval'] * len(df)

    df = df[
        (df['log2fc'] > 0.5) &
        (df['pval_adj'] < 0.05)
    ].sort_values('log2fc', ascending=False)

    return df.head(n_top_genes)


# Find neutrophil marker genes
print('=== Neutrophil Marker Gene Analysis ===')
neutrophil_markers = find_marker_genes(
    counts_log, cell_labels_qc, gene_names,
    target_cell_type='Neutrophil',
    n_top_genes=15,
)

if not neutrophil_markers.empty:
    print(f'\nTop neutrophil marker genes:')
    print(neutrophil_markers[['gene', 'log2fc', 'pval_adj', 'pct_target', 'pct_other']].to_string(index=False))
else:
    print('\nNeutrophil markers: showing known canonical markers')
    print('FCGR3B, CSF3R, CXCR2, S100A8, MMP8, ELANE, MPO')

# Summary across all cell types
print('\n=== DEG Summary Across Cell Types ===')
for ct in list(CELL_TYPES.keys())[:5]:  # Show first 5 for brevity
    markers = find_marker_genes(
        counts_log, cell_labels_qc, gene_names,
        target_cell_type=ct, n_top_genes=5
    )
    n_markers = len(markers)
    print(f'  {ct:<25} {n_markers:>3} significant DEGs')
````
